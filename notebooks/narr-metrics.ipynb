```python
import numpy as np
import matplotlib.pyplot as plt
import proplot as pplt
import seaborn as sns
import pandas as pd
import matplotlib.lines as mlines
```

```python
metrics = pd.read_csv("../data/ex1-metrics.csv")
metrics
```

```python
fig, ax = pplt.subplots(
    [1, 2],
    figsize=(4.2, 2.1),
    xlabel="Self-supervision radius (km)",
    ylabel="",
    sharey=False
)
markers = ["X", "s", "o", "^", "v"]
colors = sns.color_palette("deep", n_colors=5)
labnames = ["Accuracy", "Recall", "AIC"]
ax[0].plot(metrics.radius, metrics[["accuracy", "recall"]])
for i, c in enumerate(["accuracy", "recall"]):
    ax[0].scatter(metrics.radius, metrics[c], marker=markers[i], c=colors[i])
ax[1].plot(metrics.radius, metrics.aic, marker=markers[2], c=colors[2])
patches = [
    mlines.Line2D([], [], color=colors[i], label=labnames[i], marker=markers[i], markersize=7, linewidth=1)
    for i in range(3)
]
ax[1].legend(handles=patches, loc="ur", ncol=1, framealpha=0.5, fontsize=8)
plt.savefig("../figures/ex1-metrics.png", bbox_inches="tight")
```
